## cifar-10

```python
prob = np.load('./prob/cifar10/noise_rate_0.0/prob.npy')
print(prob.shape)
eta_corr = prob[19, :, :]
print(eta_corr.shape)
```

```python
c = 10
indices = []
```

```python
y_train = np.load('./data/cifar10/train_labels.npy')
y_train = y_train.reshape([y_train.shape[0],])
n_samples = len(y_train)
print('total number of label samples', n_samples)
```

```python
for i in range(c):
    ind = list(filter(lambda x: y_train[x] == i, range(len(y_train)))) #get the samples with true label i
    print('number of samples in this class,', len(ind)) #this is true or false vector
    #print('sum', sum(ind))
    #print(ind[1])
    eta_thresh = np.percentile(eta_corr[ind, i], 60, interpolation='higher') #get the 80th percentile acuracy
    print('class', i, 'the 60th --', eta_thresh)
    #robust_eta = eta_corr[ind, i]
    #idx_best = np.argwhere(robust_eta >= eta_thresh)
    idx_best = list(filter(lambda o: eta_corr[o, i] >= eta_thresh, ind)) #for samples with label i, if their acc>] the 80th percentile, record then
    print('total:', len(idx_best), 'max', max(idx_best))
    #idx_best = idx_best.reshape(len(idx_best))
    indices.extend(list(idx_best))   # the indices of the samples with acc >= 80% are recorded
```

```python
print(len(set(indices)))
```

```python
print(len(indices))
```

```python
print(indices)
```

```python
total = [b for b in range(len(y_train))]
```

```python
z = list(set(total) - set(indices))
```

```python
print(z)
```

```python
print(len(z))
```

```python
original_images = np.load('data/cifar10/train_images.npy')
```

```python
print(original_images.shape)
```

```python
original_labels = np.load('data/cifar10/train_labels.npy')
selected_lb = original_labels[z]
np.save('./data/cifar10/rm_40pc_labels.npy', selected_lb) #pc--percent
```

```python
selected_img = original_images[z]
np.save('./data/cifar10/rm_40pc_images.npy', selected_img) #pc--percent
```
